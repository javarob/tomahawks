```python
%matplotlib notebook
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import json
```

```python
top_wine_types_csv = "top_wine_variety_by_sales_volume.csv"
top_wine_types = pd.read_csv(top_wine_types_csv, encoding="latin")
top_wine_types.head()
```

```python
x_axis = np.arange(0,len(top_wine_types))
tick_locations = []
for x in x_axis:
    tick_locations.append(x)


plt.title("Top Wine Varieties by Sales Volumes- U.S.")  
plt.xlabel("Wine Varities")
plt.ylabel("Percentage of 2018 Sales")

#plt.xlim(-0.75, len(q2_csv_df)-0.25)
plt.ylim(ymax=20, ymin=1)

plt.figure(figsize=(15,5))
plt.bar(x_axis,(str(top_wine_types['Sales Percentage by Volume'])), color='b', alpha=0.5, align="center")
plt.xticks(tick_locations, top_wine_types['Variety'], rotation='vertical')
```
