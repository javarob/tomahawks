```python
%matplotlib notebook
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import json
plt.rcdefaults()
```

```python
q2_csv = "q2-bonded-wine-producers-by-state-1999-june-2019.csv"
q2_csv_df = pd.read_csv(q2_csv, encoding="latin")
q2_csv_df.head()
```

```python
producers_2018 = q2_csv_df['2018']
producers_2018.head()
```

```python
prod_states = q2_csv_df['State']
prod_states.head()
```

```python
#years = q2_csv_df.keys()
#columns =pd.DataFrame(years)
#columns.head()
```

```python
producers_2018_df=pd.concat([prod_states, producers_2018], axis=1)

producers_2018_df.head()
```

```python
x_axis = np.arange(0,len(producers_2018_df))
tick_locations = []
for x in x_axis:
    tick_locations.append(x)


plt.title("Wine Producers by State - U.S.")  
plt.xlabel("States")
plt.ylabel("Number of Producers")

#plt.xlim(-0.75, len(q2_csv_df)-0.25)
plt.ylim(ymax=500, ymin=1)

plt.figure(figsize=(20,10))
plt.bar(x_axis,producers_2018, color='b', alpha=0.5, align="center")
plt.xticks(tick_locations, prod_states, rotation='vertical')
```

```python
plt.rcdefaults()
fig, ax = plt.subplots()

x_pos = np.arange(len(prod_states))
ax.barh(x_pos, producers_2018, align='center',
        color='blue', ecolor='black')
ax.set_xticks(x_pos)
ax.set_xticklabels("States")
#ax.invert_yaxis()  # labels read top-to-bottom
ax.set_ylabel("Number of Producers")
ax.set_title("Wine Producers by State - U.S.")

plt.show()
```
