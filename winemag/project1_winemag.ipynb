```python
# Files to Load 
winemag1 = "../csv/winemag-data_first150k.csv"
winemag2 = "../csv/winemag-data-130k-v2.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data

winemag_data1 = pd.read_csv(winemag1)
winemag_data2 = pd.read_csv(winemag2)

# Combine the data into a single dataset

winemag_data1.count()
```

```python
#winemag_data1.columns
#winemag1_df = winemag_data1.drop([0])
winemag1_df = winemag_data1.drop(columns=["Unnamed: 0","description","region_1","region_2"])
winemag1_df.head()
```

```python
winemag2_df = winemag_data2.drop(columns=["Unnamed: 0","description","region_1","region_2","taster_name","taster_twitter_handle","title"])
winemag2_df.head()
```

```python
winemag_all_rows = pd.concat([winemag1_df, winemag2_df])
winemag_all_rows.count()
```

```python
winemag_all_rows=winemag_all_rows.reset_index(drop=True)
winemag_all_rows.head()
```

```python
by_country = winemag_all_rows.groupby("country").count()
by_country
```

```python
by_variety = winemag_all_rows.groupby("variety").count()
by_variety
```
